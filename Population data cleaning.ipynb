```python
def convert_population(value):
    if isinstance(value, str):
        if 'B' in value:
            return float(value.replace('B', '').replace(',', '')) * 1e9
        if 'M' in value:
            return float(value.replace('M', '').replace(',', '')) * 1e6
        if 'k' in value:
            return float(value.replace('k', '').replace(',', '')) * 1e3
        return float(value.replace(',', ''))
    return value 

for col in population_df.columns[1:]:
    population_df[col] = population_df[col].apply(convert_population)
```

```python
population_df
```

```python
population_df = population_df.fillna(method='ffill').fillna(method='bfill')
```

```python
total_population = population_df.describe()
print(total_population)
```
